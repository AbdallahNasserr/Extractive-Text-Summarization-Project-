# Overview of Text Summarization

Text summarization is the process of condensing long pieces of text while retaining the main points. In our project, we focused on extractive summarization, a technique that involves selecting and presenting important sentences directly from the original text.

## Techniques Used

We employed two main techniques for extractive summarization:

1. **TF-IDF (Term Frequency-Inverse Document Frequency)**: This method assigns weights to each word based on its frequency in the document and across the entire corpus.

2. **Word Embedding**: This technique represents words in a continuous vector space, allowing for semantic understanding and similarity calculation.

## Phases in the Notebook

Our project involved several phases to create text summaries:

1. **Input Data**: Gathering and loading the dataset containing the text to be summarized.

2. **Preprocessing**: Preparing the text data for analysis by performing tasks such as tokenization and stopword removal.

3. **Exploratory Data Analysis (EDA)**: Understanding the characteristics and patterns of the text data through visualization and statistical analysis.

4. **Summarization using TF-IDF**: Implementing extractive summarization using the TF-IDF technique to generate summaries based on word importance.

5. **Summarization using Word Embedding**: Employing word embedding techniques to further enhance the summarization process by capturing semantic relationships between words.

By following these phases, we were able to gain insights into the dataset, preprocess the text effectively, and generate concise summaries using both TF-IDF and word embedding techniques.

```python
import seaborn as sns


word_list = words

similarity_matrix = np.zeros((len(word_list), len(word_list)))
for i, word1 in enumerate(word_list):
    for j, word2 in enumerate(word_list):
        if i != j:
            similarity_matrix[i, j] = nlp(word1).similarity(nlp(word2))

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, annot=True, xticklabels=word_list, yticklabels=word_list, cmap="YlGnBu")
plt.title("Word Similarity Heatmap")
plt.xlabel("Words")
plt.ylabel("Words")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

In TF-IDF Summarization, the primary goal is to condense a large amount of text into a shorter summary while retaining the most important information.

## Methodology

- **Term Frequency-Inverse Document Frequency (TF-IDF)**: TF-IDF is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents. It considers both the frequency of the word in the document (TF) and the rarity of the word in the entire corpus of documents (key concepts.

## Benefits

- **Efficiency**: TF-IDF summarization can efficiently condense large volumes of text into shorter summaries, making it easier to extract relevant informatmation retrieval tasks.


```python
def remove_special_characters(text):
    regex = r'[^a-zA-Z0-9\s]'
    text = re.sub(regex,'',text)
    return text
```

```python
def freq(words):
    words = [word.lower() for word in words]
    freqTable = {}
    words_unique = []
    for word in words:
        if word not in words_unique:
            words_unique.append(word)
    for word in words_unique:
        freqTable[word] = words.count(word)
    return freqTable
```

```python
def tf_score(word,sentence):
    freq_sum = 0
    word_frequency_in_sentence = 0
    len_sentence = len(sentence)
    for word_in_sentence in sentence.split():
        if word == word_in_sentence:
            word_frequency_in_sentence = word_frequency_in_sentence + 1
    tf =  word_frequency_in_sentence/ len_sentence
    return tf
```

```python
def idf_score(no_of_sentences,word,sentences):
    no_of_sentence_containing_word = 0
    for sentence in sentences:
        sentence = remove_special_characters(str(sentence))
        sentence = re.sub(r'\d+', '', sentence)
        sentence = sentence.split()
        sentence = [word for word in sentence if word.lower() not in Stopwords and len(word)>1]
        sentence = [word.lower() for word in sentence]
        sentence = [wordlemmatizer.lemmatize(word) for word in sentence]
        if word in sentence:
            no_of_sentence_containing_word = no_of_sentence_containing_word + 1
    idf = math.log10(no_of_sentences/no_of_sentence_containing_word)
    return idf
```

```python
def pos_tagging(text):
    pos_tag = nltk.pos_tag(text.split())
    pos_tagged_noun_verb = []
    for word,tag in pos_tag:
        if tag == "NN" or tag == "NNP" or tag == "NNS" or tag == "VB" or tag == "VBD" or tag == "VBG" or tag == "VBN" or tag == "VBP" or tag == "VBZ":
            pos_tagged_noun_verb.append(word)
    return pos_tagged_noun_verb
```

```python
def tf_idf_score(tf,idf):
    return tf*idf
```

```python
def word_tfidf(freqTable,word,sentences,sentence):
    word_tfidf = []
    tf = tf_score(word,sentence)
    idf = idf_score(len(sentences),word,sentences)
    tf_idf = tf_idf_score(tf,idf)
    return tf_idf
```

```python
def sentence_importance(sentence,freqTable,sentences):
    sentence_score = 0
    sentence = remove_special_characters(str(sentence))
    sentence = re.sub(r'\d+', '', sentence)
    pos_tagged_sentence = []
    no_of_sentences = len(sentences)
    pos_tagged_sentence = pos_tagging(sentence)
    for word in pos_tagged_sentence:
        if word.lower() not in Stopwords and word not in Stopwords and len(word)>1:
            word = word.lower()
            word = wordlemmatizer.lemmatize(word)
            sentence_score = sentence_score + word_tfidf(freqTable,word,sentences,sentence)
    return sentence_score
```

# Word Embedding Summarizer (Word2Vec)

Word Embedding Summarization, using Word2Vec, generates summaries by understanding the semantic relationships between words.

## Methodology: 
Converts words into vectors using Word2Vec models to capture semantic similarities.
## Features: 
Ranks sentences based on semantic similarity, extracts keywords, and phrases.
## Application:
Produces accurate summaries and extracts key concepts from text.

Word Embedding Summarizer offers a powerful and accurate approach to automatic text summarization.

```python
# Extract word vectors
word_embeddings = {}
f = open(glove_input_file, encoding='utf-8')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    word_embeddings[word] = coefs
f.close()
```
